В матрице отчетливо видна как положительная корреляция между переменными, так и отрицательная

```python
indices = np.where((df.corr() >= 0.4) & (df.corr() < 1))
cols = set()
for x, y in zip(*indices):
    cols.add(df.corr().index[x])
    cols.add(df.corr().columns[y])
```

```python
cols
```

```python
cols = ['REG_FACT_FL', 
        'FACT_POST_FL', 
        'REG_POST_FL', 
        'REG_FACT_POST_FL', 
        'OWN_AUTO', 
        'AUTO_RUS_FL', 
        'CREDIT', 
        'TERM', 
        'FST_PAYMENT', 
        'LOAN_NUM_TOTAL', 
        'LOAN_NUM_CLOSED', 
        'LOAN_NUM_PAYM', 
        'LOAN_DLQ_NUM', 
        'LOAN_MAX_DLQ', 
        'LOAN_AVG_DLQ_AMT', 
        'LOAN_MAX_DLQ_AMT']
```

```python
size = len(cols)
cm = np.corrcoef(df[cols].values.T)
a4_dims = (11, 8)
_, ax = plt.subplots(figsize=(16, 16))
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, ax=ax, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': size}, yticklabels=cols, xticklabels=cols)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()
```

Корреляция между переменными:
1. REG_FACT_FL, FACT_POST_FL, REG_POST_FL, REG_FACT_POST_FL - адрес
2. OWN_AUTO, AUTO_RUS_FL - автомобиль
3. CREDIT, TERM, FST_PAYMENT - кредит
4. LOAN_NUM_TOTAL, LOAN_NUM_CLOSED, LOAN_NUM_PAYM - ссуды
5. LOAN_DLQ_NUM, LOAN_MAX_DLQ, LOAN_AVG_DLQ_AMT, LOAN_MAX_DLQ_AMT - просрочки

```python
indices = np.where(df.corr() <= -0.3)
cols = set()
for x, y in zip(*indices):
    cols.add(df.corr().index[x])
    cols.add(df.corr().columns[y])
```

```python
size = len(cols)
cm = np.corrcoef(df[cols].values.T)
a4_dims = (11.7, 8.27)
_, ax = plt.subplots(figsize=(7, 7))
sns.set(font_scale=1)
hm = sns.heatmap(cm, ax=ax, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': size}, yticklabels=cols, xticklabels=cols)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()
```
